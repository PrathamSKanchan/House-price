````python
X = data[['Owner Income', 'House Age', 'Number of Rooms', 'Number of Bedrooms', 'Area Population']]
y = data['Price']
````

````python
from sklearn.model_selection import train_test_split
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=100)
````

````python
X_test.info()
````

````python
X_train.info()
````

````python
from sklearn.linear_model import LinearRegression
````

````python
lin = LinearRegression()
````

````python
lin.fit(X_train,y_train)
````

````python
intercept = lin.intercept_
````

````python
coefficient = lin.coef_
````

````python
print(intercept, coefficient)
````

````python
X_train.columns
````

````python
matrix = pd.Series(coefficient, X_train.columns)
````

````python
matrix
````

````python
intercept
````

````python
predictions = lin.predict(X_test)
````

````python
sns.scatterplot(x=predictions,y = y_test)
````

````python
sns.scatterplot(x=y_test,y=y_test)
````

````python
from sklearn.metrics import mean_absolute_error,mean_squared_error
````

````python
MAE = mean_absolute_error(y_test,predictions)
MSE = mean_squared_error(y_test,predictions)
RMSE = np.sqrt(MSE)
````

````python
print(str(MAE)+'|'+ str(MSE)+'|'+ str(RMSE)+ '|' + str((RMSE / Mean)*100) )
````

````python
a = lin.coef_
a
````

````python
b = lin.intercept_
b
````

````python
def HousePrice(OwnerIncome, HouseAge, NumberofRooms, NumberofBedrooms, AreaPopulation):
  price = OwnerIncome*a[0] + HouseAge*a[1] + NumberofRooms*a[2] + NumberofBedrooms*a[3] + AreaPopulation*a[4] + b
  return price
````

````python
print('House price is expected to be:')
HousePrice(556815, 5.68, 7, 4, 23087)
````

````python
print('House price is expected to be:')
HousePrice(548758, 6.99, 7, 2, 36516)
````
